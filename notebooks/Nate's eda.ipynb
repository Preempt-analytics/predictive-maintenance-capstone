```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Read the product_art.csv file in and assign in to a df
df = df=pd.read_csv("../data/ai4i2020.csv")




df.head(10)


```

```python
#initial EDA
df.info()

df.describe()
```

```python
# quick overview plot
fig = px.histogram(df, x="Type", color="Machine failure", barmode="group")
fig.update_layout(title="Machine Failure by Type", xaxis_title="Type", yaxis_title="Count")
fig.show()
```

```python
# Correlation matrix heatmap
import plotly.figure_factory as ff

correlation_matrix = df.corr(numeric_only=True)
fig_corr = ff.create_annotated_heatmap(z=correlation_matrix.values, x=correlation_matrix.columns, y=correlation_matrix.columns)
fig_corr.update_layout(title="Correlation Matrix")
fig_corr.show()

# Distribution of numeric columns
fig_dist = px.histogram(df, nbins=30, title="Distribution of Numeric Columns")
fig_dist.show()

# Machine failure rate
failure_rate = df['Machine failure'].value_counts()
fig_pie = px.pie(values=failure_rate.values, names=['No Failure', 'Failure'], title="Machine Failure Rate")
fig_pie.show()
```
